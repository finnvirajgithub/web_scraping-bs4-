## HTTP error
### The page is not found on the server

```python
from urllib.request import urlopen
from urllib.error import HTTPError
try:
    html = urlopen('https://www.pythonscraping.com/pages/page1.html')
except HTTPError as e:
    print('The page is not found on the server')
    
else:
    print('It worked')
```

## URL error
### The server is not found

```python
from urllib.error import URLError

try:
    html = urlopen('https://www.pythonscraping.com/pages/page1.html')
except URLError as e:
    print('The server could not be found')
else:
    print('It worked')
    
```

## Final code

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        bs= BeautifulSoup(html.read(),'html.parser')
        title = bs.h1
    except AtrributeError as e:
        return None
    return title

title = getTitle('https://www.pythonscraping.com/pages/page1.html')
if title == None:
    print('Title could not found')
else:
    print(title)
```
